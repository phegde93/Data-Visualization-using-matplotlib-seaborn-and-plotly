## Histogram

```python
plt.title("Distribution of Sepal Width")
plt.hist(x="sepal_width", data=flowers_df);
```

## Multiple Histogram

```python
flowers_df["species"].unique()
```

```python
filt_setosa = flowers_df["species"] == "setosa"
filt_versi = flowers_df["species"] == "versicolor"
filt_vigin = flowers_df["species"] == "virginica"
```

```python
setosa_df = flowers_df.loc[filt_setosa,]
versi_df = flowers_df.loc[filt_versi,]
virgin_df = flowers_df.loc[filt_vigin,]
```

```python
plt.hist("sepal_width", data=setosa_df, alpha=0.4)
plt.hist("sepal_width", data=versi_df, alpha=0.4);
plt.title("Distribution of setosa and Versicolor")
plt.xlabel("Sepal Width")
plt.ylabel("Count")
plt.legend(["Setosa","Versicolor"]);
```

## Stacked Histogram

```python
plt.title("Distribution of Sepal Width")
plt.hist([setosa_df["sepal_width"],versi_df["sepal_width"], virgin_df["sepal_width"]],stacked=True)
plt.legend(["Setosa","Versicolor","Virginica"]);
```

### Plotting Multiple chart in a grid

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Use the axes for plotting
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')


# Pass the axes into seaborn
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0,1]);

# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versi_df.sepal_width, virgin_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df_pivot, cmap='Blues', ax=axes[1,1]);

# Plot an image using the axes
axes[1,2].set_title('Bar Chart')
axes[1,2].bar(years,apples)
axes[1,2].bar(years,oranges, bottom=apples)
axes[1,2].legend(["Apples","Oranges"])


plt.tight_layout(pad=2);
```

```python
import plotly.express as px
```

```python
# Let's downlaod world-wise population dataset
```

```python
population_csv_url = 'https://gist.githubusercontent.com/aakashns/bbd36fbd7c0be266f0c875ad2006a9fd/raw/1763ee47c8919995c4115bb063c99511ced34712/population.csv'
population_df = pd.read_csv(population_csv_url, index_col='Year')
```

```python
population_df.head()
```

## Interactive Line chart

```python
show =px.line(population_df["Hungary"], title="Population of Hungary");
```

```python
show
```

```python
europe_df = population_df[['Hungary', 'Czech Republic', 'Switzerland']]
europe_df.head()
```

```python
fig = px.line(europe_df, title="Population", color_discrete_sequence=["aquamarine", "cornflowerblue", "goldenrod"])
fig.update_layout(font_size=14, legend_title="Country")
fig.update_yaxes(rangemode="tozero")
```

## Interactive Bar Chart

```python
fig =px.bar(population_df[["India","Pakistan"]])
fig.update_layout(title="Population",legend_title="Country")
```

```python
population_i_c = population_df.loc[:,["India","China"]]
```

```python
population_diff = population_i_c.diff()
```

```python
population_diff
```

```python
fig =px.line(population_diff, title="Change in Population per year",
             color_discrete_sequence=["aquamarine", "cornflowerblue"])
fig.update_layout(font_size=14, legend_title="Country")
```

## Creating 3-D graphs

```python
import pandas as pd

z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv', index_col=0)
z_data.head()
```

```python
import plotly.graph_objects as go
```

```python
surface = go.Surface(z=z_data.values)
fig = go.Figure(surface)
fig.update_layout(title='Mt. Bruno Elevation')
fig.show()
```

```python
fig=px.scatter_3d(iris_df,
             x="sepal_length",
             y="sepal_width",
             z="petal_width",
             color="species",
                 title="Iris Flower")
fig.show()
```
